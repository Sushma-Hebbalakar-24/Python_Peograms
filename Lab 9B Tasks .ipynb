# Visualization Titanic  Samples

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Replace 'titanic_data.csv' with the path to your dataset
file_path = 'E:\\NCI\\Database & Analytics Programming\\Week 4\\train.csv'

# Read the data into a DataFrame
titanic_data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to check if the data is loaded correctly
titanic_data
```

# Visualization 1: Survival Scatter Plot (Scatter Plot, X-axis=Pclass, Y-axis=Fare, Hue=Survived)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with columns 'Pclass', 'Fare', and 'Survived'

# Sample data (replace this with your actual dataset)
import pandas as pd

# Create a survival scatter plot
sns.scatterplot(x='Pclass', y='Fare', hue='Survived', data=titanic_data, palette='Set2')

# Add labels and title
plt.title('Survival Scatter Plot')
plt.xlabel('Pclass')
plt.ylabel('Fare')

# Show the plot
plt.show()
```

# Visualization 2: Survival Bar Chart by Gender

```python
sns.countplot(x='Sex', hue='Survived',data=titanic_data)

# Add labels and title
plt.title('Survival Bar Chart')
plt.xlabel('Gender')
plt.ylabel('Survival')

# Show the plot
plt.show()
```

# Visualization 3: Correlation Heatmap

```python
#df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = titanic_data.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Visualization 4: Age Distribution by Survival

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', hue='Survived', data=titanic_data,element='step',stat='density', common_norm=False,kde=True)
plt.title('Age Distribution by Survival')
plt.show()
```

# Visualization 5: Fare Distribution by Class

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=titanic_data, palette='Pastel2')
plt.title('Fare Distribution by Class')
plt.show()
```

# Visualization 6: Embarked Port Pie Chart

```python
titanic_data['Embarked'].value_counts().plot.pie( autopct='%1.1f%%', startangle=90 )
```

# Visualization 7: Family Size vs. Survival Bubble Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fare', y='Age', size='Survived', data=titanic_data, sizes=(20, 200), palette='viridis')
plt.title('Family Size vs. Survival Bubble Plot')
plt.show()
```

# Visualization 8: Passenger Class Stacked Bar Chart

```python
stacked_data = titanic_data.groupby(['Pclass', 'Survived']).size().unstack()
stacked_data.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])

# Add labels and title
plt.title('Passenger Class Stacked Bar Chart')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])

# Show the plot
plt.show()
```
