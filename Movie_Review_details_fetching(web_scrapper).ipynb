```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

#Send a Get request to the IMDB top movies page
url ="https://timesofindia.indiatimes.com/entertainment/kannada/movie-reviews/"
response=requests.get(url)

#Parse the HTML content using BeautifulSoup
soup= BeautifulSoup(response.text,'html.parser')
soup
#Extract movie data

movies=[]

for movie in soup.find_all('div',class_='FIL_right'):
    title = movie.a.text
    year = movie.div.text.strip('()')
#year = movie.span.text.strip('()')
    rating = movie.find_next('span',class_='star_count')
    movies.append([title,year,rating])
# Create a DataFrame from the extracted data
df = pd.DataFrame(movies, columns=['Title','Year','Rating'])
# Save the data to a CSV file
df.to_csv('top_movies.csv', index=False)
print(df)

```
