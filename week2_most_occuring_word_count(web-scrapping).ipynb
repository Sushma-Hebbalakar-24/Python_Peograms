# In Python, you can use the `collections.Counter` class to find the most common elements in an iterable, such as a list. The `most_common()` method of the `Counter` class is used to accomplish this. Here's how you can use it:

```python

````python

from collections import Counter

# Sample list of elements
elements = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# Create a Counter object to count the occurrences of each element
element_counter = Counter(elements)

# Use most_common() to get the most common elements and their counts
most_common_elements = element_counter.most_common()

# Print the result
print(most_common_elements)
```

````

In this example:

1. We import the `Counter` class from the `collections` module.

2. We create a list called `elements` with some sample data.

3. We create a `Counter` object called `element_counter` by passing the `elements` list to it. This counts the occurrences of each element in the list.

4. We use the `most_common()` method on the `element_counter` object to retrieve a list of tuples where each tuple contains an element and its count, sorted by count in descending order.

5. Finally, we print the `most_common_elements`, which will display the elements in the order of their frequency, with the most common elements appearing first.

You can adapt this code to find the most common non-numeric values in a list of strings or any other iterable.

````python
!pip install wordcloud
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Step 1: Web Scraping
url = "https://realpython.com/documenting-python-code/"  # Replace with the URL of the webpage you want to scrape
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text()

# Step 2: Text Preprocessing (you may need more preprocessing depending on your data)
# Remove punctuation and convert to lowercase
text = ''.join([char.lower() for char in text if char.isalnum() or char.isspace()])

# Step 3: Word Frequency Calculation
word_frequencies = {}
words = text.split()
for word in words:
    if word not in word_frequencies:
        word_frequencies[word] = 1
    else:
        word_frequencies[word] += 1

# Step 4: Word Cloud Generation
wordcloud = WordCloud(width=800, height=400).generate_from_frequencies(word_frequencies)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
````

# Creating Word Cloud using particular Image as pattern

````python
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Step 1: Web Scraping
url = "https://realpython.com/documenting-python-code/"  # Replace with the URL of the webpage you want to scrape
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text()

# Step 2: Text Preprocessing (you may need more preprocessing depending on your data)
# Remove punctuation and convert to lowercase
text = ''.join([char.lower() for char in text if char.isalnum() or char.isspace()])

# Step 3: Word Frequency Calculation
word_frequencies = {}
words = text.split()
for word in words:
    if word not in word_frequencies:
        word_frequencies[word] = 1
    else:
        word_frequencies[word] += 1       

# Load the custom image
custom_image = np.array(Image.open('C:\\Users\\User\\Downloads\\clud2.jpg'))

# Create a WordCloud object with the custom image as the mask
wordcloud = WordCloud(width=1000, height=600, background_color='white', mask=custom_image).generate_from_frequencies(word_frequencies)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
````
