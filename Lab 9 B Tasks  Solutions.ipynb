# Visualization Titanic  Samples

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Replace 'titanic_data.csv' with the path to your dataset
file_path = 'C:\\Users\knowl\\Downloads\\train (1).csv'

# Read the data into a DataFrame
titanic_data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to check if the data is loaded correctly
titanic_data.head()
```

# Visualization 1: Survival Scatter Plot (Scatter Plot, X-axis=Pclass, Y-axis=Fare, Hue=Survived)

```python

# Set the style for Seaborn plots
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Pclass', y='Fare', hue='Survived', data=titanic_data, palette='Set1')
plt.title('Survival Scatter Plot')
plt.show()
```

# Visualization 2: Survival Bar Chart by Gender

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=titanic_data, palette='Set2')
plt.title('Survival Bar Chart by Gender')
plt.show()
```

# Visualization 3: Correlation Heatmap

```python
Corrdata= titanic_data[['Survived','Age','SibSp','Parch','Fare']]
plt.figure(figsize=(10, 6))
correlation_matrix = Corrdata.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Visualization 4: Age Distribution by Survival

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', hue='Survived', data=titanic_data, element='step', stat='density', common_norm=False, kde=True)
plt.title('Age Distribution by Survival')
plt.show()
```

# Visualization 5: Fare Distribution by Class

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=titanic_data, palette='Pastel1')
plt.title('Fare Distribution by Class')
plt.show()
```

# Visualization 6: Embarked Port Pie Chart

```python
plt.figure(figsize=(8, 8))
titanic_data['Embarked'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.title('Embarked Port Pie Chart')
plt.show()
```

# Visualization 7: Family Size vs. Survival Bubble Plot

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fare', y='Age', size='Survived', data=titanic_data, sizes=(20, 200), palette='viridis')
plt.title('Family Size vs. Survival Bubble Plot')
plt.show()
```

# Visualization 8: Passenger Class Stacked Bar Chart

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=titanic_data, palette='Accent')
plt.title('Passenger Class Stacked Bar Chart')
plt.show()
```

# Visualization 9: Gender vs. Age Violin Plot

```python

plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Age', data=titanic_data, inner='quartile', palette='Set3')
plt.title('Gender vs. Age Violin Plot')
plt.show()
```

# Visualization 10: Cabin Deck Bar Chart

```python

plt.figure(figsize=(10, 6))
titanic_data['Survived'] = titanic_data['Survived'].fillna('Unknown')
sns.countplot(x='Survived', data=titanic_data, palette='Set1', order=titanic_data['Survived'].value_counts().index)
plt.title('Cabin Deck Bar Chart')
plt.show()
```
